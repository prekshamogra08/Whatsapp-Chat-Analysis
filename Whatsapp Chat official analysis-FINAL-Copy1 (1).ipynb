```python
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
df.message = df.message.apply(lambda x: remove_emoji(x))
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
sorted(list(enumerate(similarity[0])), reverse = True, key=lambda x:x[1])[1:10]
```

```python
def recommend(name):
    name_index = df[df.user==name].index[0]
    distances = similarity[name_index]
    name_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:3]
    
    for i in name_list:
        print (df.iloc[i[0]].user)
```

```python
original_vals = df_new.new_column.copy()

# get max absolute value
original_max = max([abs(val) for val in original_vals])

# normalize to desired range size
new_range_val = 1
df_new.new_column = [float(val)/original_max * new_range_val for val in original_vals]
```

```python
nature = []
for i in df_new.new_column:
    if -0.5<=i<0:
        nature.append("Sad")
    elif -1<=i<-0.5:
        nature.append("Melancholic :( ")
    elif i==0:
        nature.append("Neutral")
    elif 0<i<=0.5:
        nature.append("Happy")
    elif 0.5<i<=1:
        nature.append("Ecstatic :) ")
```

```python
df_new['Nature']=nature
```

```python
cols = ['new_column'] # one or more

Q1 = df_new[cols].quantile(0.25)
Q3 = df_new[cols].quantile(0.75)
IQR = Q3 - Q1

df_new = df_new[~((df_new[cols] < (Q1 - 1.5 * IQR)) |(df_new[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```
